```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing GME revenue data
url = 'https://companiesmarketcap.com/inr/gamestop/revenue/'

# Send a GET request to the website
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Example: Find the table containing revenue data (adjust based on actual HTML structure)
table = soup.find('table')  # Adjust selector based on the actual HTML

if table:
    # Extract table headers
    headers = [th.text for th in table.find_all('th')]

    # Extract table rows
    rows = []
    for row in table.find_all('tr')[1:]:  # Skip header row
        cols = [td.text for td in row.find_all('td')]
        rows.append(cols)

    # Create a DataFrame
    gme_revenue = pd.DataFrame(rows, columns=headers)

    # Display the last five rows
    print(gme_revenue.tail())
else:
    print("Table not found. Check the HTML structure and update the selector.")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to plot the stock data
def make_graph(df, title):
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['Close'], label='Closing Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2024-01-01', end='2024-09-01')

# Plot the data
make_graph(tesla_data, 'Tesla Stock Price (2024)')
```
